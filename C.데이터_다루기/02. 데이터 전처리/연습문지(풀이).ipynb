문제1 
네 종류의 식단 A1, A2, A3, A4에 따른 식이용법이 혈액응고시간에 영향을 주는지 알고자 실험한 자료가 다음과 같다. 
인자의 수준을 나타내는 변수를 class로, 실습값을 time으로 하여 데이터를 구성하여라.

-> 변수 = column


```python
식단 = DataFrame({'class':['A1','A2','A3','A4'],'TIME':[[62,60,63,59,61],[62,67,71,64,65,66],[68,66,71,67,68,69,70],[58,62,60,59,61,63,64]]})
식단
```

```python
식이요법 = {'A1':[62,60,63,59,61],
        'A2':[62,67,71,64,65,66],
        'A3':[68,66,71,67,68,69,70],
        'A4':[58,62,60,59,61,63,64]}
reuslt = []
for i in 식이요법 :
    for j in range(len(식이요법[i])) : 
        reuslt.append([i,식이요법[i][j]])
df = DataFrame(reuslt,columns=['CLASS','TIME'])
df
```

```python
# 생성된 데이터 프레임에 대한 각 변수의 데이터 타입 확인
df.dtypes

# object를 변경해야 해야지 카테고리로~ (내부적으로 데이터를 적게 잡아서 연산 속도가 빨라짐 )
```

```python
# 명목척도 
# 타입 변경 후 복제본을 만드는 방법
df2 = df.astype({'CLASS':'category'})
df2.dtypes

# 원본에 바로 저장하려면 
df['CLASS'] = df['CLASS'].astype({'CLASS':'category'})
```

```python
# 서열척도
from pandas import CategoricalDtype
df['CLASS'] = df['CLASS'].astype(CategoricalDtype(categories=['A1','A2','A3','A4'],ordered=True)) # 서열이 생긴다. a1 -> a4 

df.dtypes
```

```python
식단.loc[0] 
# loc, iloc integer location의 약자로 데이터프레임의 행이나 칼럼의 순서를 나타내는 정수로 특정 값을 추출해오는 방법이다. loc는 컬럼명을 직접 적거나 특정 조건식을 써줌으로써 사람이 읽기 좋은 방법으로 데이터에 접근하는 방법이며, iloc는 컴퓨터가 읽기 좋은 방법(숫자)로 데이터가 있는 위치에 접근한다. df.iloc[행인덱스,열인덱스] -> 전체 데이터프레임에서 인덱스 이름이 0인 행만 추출해줘 
```

```python
result=[]
for a in 식단.index: # 4번 반복한다. 
    for x in 식단.loc[a].to_list():  # A1 [62, 60, 63, 59, 61]
        result.append(list(Series([a,x],index=["class",'time']))) #[0, A1 [62, 60, 63, 59, 61]] 
print(result)
df = DataFrame(df,columns=['CLASS','TIME'])
df
```

문제3 
다음은 한 도시의 20개 중,고교를 대상으로 조사한 흡연율 자료이다. 도시와 농촌별에 따른 남여 평균 흡연율을 확인할 수 있는 데이터 피벗테이블로 구성하시오.

```python
df = read_excel('https://data.hossam.kr/test/q3.xlsx')
df
```

```python
pivot_table(df,                        # 피벗할 데이터프레임
            index = '지역',             # 행 위치에 들어갈 열
            columns = '구성',           # 열 위치에 들어갈 열
            values = '흡연율',            # 데이터로 사용할 열
            aggfunc =  'mean'     # 데이터 집계함수
)
```
