## 병합대상 열 지정하기

```python
고객가입 = [['지디','2018-01-01',1000000],['태양','2019-01-01',20000,]]
df_l = DataFrame(고객가입,columns=['고객명','날짜','데이터'])
df_l
```

```python
고객나이 = [['지디','21세'],['태양','22세']]
df_r = DataFrame(고객나이,columns=['고객명','데이터'])
df_r
```

```python
temp = merge(df_l,df_r,on='고객명') # merge(데이터프레임,데이터프레임,병합기준 열) 
temp #병합 기준 열이 아니면서 이름이 같은 열에는 _x 또는 _y 와 같은 접미사가 붙는다. 
```

```python
# 컬럼명 변경은 rename 
temp.rename(columns={'데이터_x':'금액','데이터_y':'나이'})
```

#### 같은 내용인데, 컬럼명이 다른 경우

- 변수 = merge(데이터프레임,데이터프레임,left_on=[컬럼],right_on=[컬럼])
- 변수.drop('지울 컬럼명',axis=1) 

```python
df_left3 = DataFrame({'이름':['영희','철수'], '국어':[87,91]})
df_right3 = DataFrame({'성명':['영희','철수'],'영어':[90,82]})
# 한쪽은 이름이고 한쪽은 성명이라고 씀.
```

```python
r3 = merge(df_left3, df_right3, left_on=['이름'],right_on=['성명'])
r3
# 그냥 다 붙음 
```

```python
r4 = r3.drop('성명',axis=1) 
r4 
# 붙인것에서 성명 변수를 지워라 
```

### 병합 - 인덱스를 활용

```python
# 학생의 이름을 인덱스로 갖는 두 데이터 프레임
df_left = DataFrame({'수학':[90,82]}, index=['호열','구식'])
df_right=DataFrame({'국어':[90,82]}, index=['호열','대남'])

merge(df_left,df_right,left_index=True, right_index=True) 
# inner join 과 같아서 양쪽 인덱스를 다 가진 호열이만 살아남음 
```

```python
merge(df_left,df_right,left_index=True,right_index=True,how='outer') # full join과 같다
```

인덱스가 0,1.. 과 인덱스가 사람이름인 데이터프레임 합치기

```python
df1 = DataFrame({'수학':[90,82]},index=['낙수','동오'])
df2 = DataFrame({'성명':['낙수','성구'],'영어': [87,90]})
```

```python
tem = merge(df_left,df_right,left_index=True,right_on=['성명'],how='outer') 
tem
```

```python
tmp2 = tem.set_index('성명') # 성명을 지운다. 
tmp2
```
