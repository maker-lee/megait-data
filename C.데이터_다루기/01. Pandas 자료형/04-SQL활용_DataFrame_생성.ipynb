테이블의 각 record를 튜플로 표현하는 리스트 객체를 얻을 수 잇다. 

```python
# 데이터 조회를 위한 커서 객체 생성 
cursor = dbcon.cursor()
```

```python
# 데이터 조회를 위한 sql 처리
sql = 'SELECT * FROM board_main_post'
cursor.execute(sql) 
result = cursor.fetchall()
result
```

```python
# 딕셔너리 형태로 데이터 조회
# 데이터 조회를 위한 커서 객체 생성
# 객체 생성 시 파라미터를 전달해야 한다.
cursor = dbcon.cursor(pymysql.cursors.DictCursor)
```

```python
# 데이터 조회하기
sql = 'SELECT * FROM board_main_post'
cursor.execute(sql) 
result = cursor.fetchall()
result
```

```python
# 조회 결과를 데이터 프레임으로 변환
df = DataFrame(result)
df
```

```python
# 데이터 프레임에 대한 인덱스 설정
df.set_index('id',inplace=True)
df
```

```python
# 입력, 수정, 삭제 수행 방식은 동일. 데이터 조회 과정에서 생성한 cursor 객체 재사용
```

```python
# 데이터입력
sql = '''INSERT INTO board_main_post (title, contents,created_at, updated_at) VALUES ('pandas test','테스트 중입니다.',now(),now())'''

rows = cursor.execute(sql)
print('%d개의 행이 저장됨' % rows)
print('생성된 pk: %d' % cursor.lastrowid)

# 처리 결과를 실제로 반영함
dbcon.commit()

# dbcon.rollback()으로 되돌리기
```

```python
# 데이터 수정
sql = '''UPDATE board_main_post SET 
title ='수정된 제목',
contents = '수정된 내용',
updated_at = now() 
WHERE id = 22'''
print(sql)
rows = cursor.execute(sql)
print('%d개의 행이 갱신됨' % rows)
# 처리 결과를 실제로 반영함
dbcon.commit()
```

```python
# 데이터 삭제
sql = 'DELETE FROM board_main_post WHERE id > 15'
rows = cursor.execute(sql)
print('%d개의 행이 삭제됨' % rows)
dbcon.commit()
```

```python
# 데이터 베이스 접속 해제
cursor.close()
dbcon.close()
```

## 02) SQLAlchemy 사용

1) 데이터 베이스 접속 

```python
# 데이터베이스 접속 패키지 설치
pymysql.install_as_MySQLdb() #pip이 없음 
import MySQLdb
```

```python
# 접속 문자열 생성
# mysql+mysqldb://계정이름:비밀번호@:포트번호/데이터베이스이름?charset=인코딩
conStr = 'mysql+mysqldb://root:1234@:3306/board?charset=utf8'
conStr
```

```python
# 데이터베이스 접속하기
engine = create_engine(conStr)
conn = engine.connect() # conn ; 접속객체
```

```python
# 데이터 조회하기
# sql문 사용하기
df = read_sql('SELECT * FROM board_main_post',index_col='id',con=conn)
df
```

```python
# 테이블의 데이터를 직접 가져오기
# 모든 데이터 조회
df = read_sql_table('board_main_post',con=conn) # 편한데 where문 불가 
df
```

```python
# 인덱스를 지정한 조회
# read_sql_table 함수를 사용할 경우 where 절은 사용할 수 없다. 
df = read_sql_table('board_main_post',index_col='id',con=conn)
df
```

```python
# 특정 칼럼만 가져오기
df = read_sql_table('board_main_post',index_col='id',columns=['title','contents'],con=conn)
df
```

```python
# 데이터 내보내기
df.to_sql(name='new_table',con=conn,if_exists='append',index=False)
conn.commit()

# append : 이어쓰기
# replace ; 새로쓰기
```

name = '테이블명' 이름으로 기존 테이블이 있으면 해당 테이블의 컬럼명에 맞게 데이터를 넣을 수 있다.
if_exist = 'append' 옵션이 있으면, 기존 테이블에 데이터를 추가로 넣는다.
if_exist = 'fail' 옵션이 있으면, 기존 테이블이 있을 경우 아무 일도 하지 않음
if_exist = 'replace' 옵션이 있으면 기존 테이블이 있을 경우 기존 테이블을 삭제 하고 다시테이블을 만들어서 새로 데이터를 넣고, 이미 만들어진 테이블이엇으면 name= '테이블명' 이름으로 테이블을 자동으로 만들고 데이너를 넣을 수 있다. 테이블이 자동으로 만들어지므로 구조가 최적화 되지 않아 테이블 만드는 것은 추천하지 않는다.
