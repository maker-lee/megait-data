```python
import pymysql
from sqlalchemy import create_engine
from pandas import DataFrame
from pandas import read_sql, read_sql_table
from pandas import merge
```

```python
# 데이터베이스 접속
dbcon = pymysql.connect(host='localhost',
                        port=3306,
                        user='root',
                        password='1234',
                        db = 'semiproject',
                        charset='utf8')
# 데이터 조회하기
cursor = dbcon.cursor(pymysql.cursors.DictCursor)
```

```python
# 다음검색 데이터 조회를 위한 sql 처리
sql = 'SELECT * FROM daum'
cursor.execute(sql) 
result = cursor.fetchall()
# 조회 결과를 데이터 프레임으로 변환
df = DataFrame(result)
df
```

```python
# 딕셔너리 형태로 데이터 조회
# 데이터 조회를 위한 커서 객체 생성
# 객체 생성 시 파라미터를 전달해야 한다.
cursor = dbcon.cursor(pymysql.cursors.DictCursor)
```

```python
# 요기요 데이터 조회하기
sql = 'SELECT * FROM yogiyo'
cursor.execute(sql) 
result1 = cursor.fetchall()
result1

# 조회 결과를 데이터 프레임으로 변환
df2 = DataFrame(result1)
df2
```

```python
new = merge(left = df2, right= df, how = 'left', on='매장명')
new
```

```python
# 접속 문자열 생성
conStr = 'mysql+mysqldb://root:1234@:3306/semiproject?charset=utf8' # utf8db 디비명 
engine = create_engine(conStr)
conn = engine.connect() # conn ; 접속객체

# 데이터 내보내기
new.to_sql(name='merge',con=conn,if_exists='append',index=False)
conn.commit()
new
```
