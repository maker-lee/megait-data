```python
import requests
import json
from pandas import DataFrame # 딕셔너리를 포함한 리스트를 표 형태로 표기함
```

```python
# URL 기본 형식
# query 파라미터 : query는 변수 이름, 검색어 겟 방식이므로 ?query=검색어  그 외 더 찾고 싶은게 있다면 ?query=검색어&sort=latest?query=검색어&size=50 이런 방식으로 입력
urlFmt = "https://dapi.kakao.com/v3/search/book?query={query}&page={page}&size={size}"
```

```python
# 요청변수
query = "파이썬"
page = 1
size = 50
key = "87fa270de879f3dd014b8779623884db" # 카카오 개발자에서 발급받은 API 연동키
```

```python
# 접속 객체 생성
session = requests.Session()

# 접속 정보 설정 # 오픈 api는 누구나 가져가라는 의미이므로 Referer가 필요하지 않다. 
# 검색>Daum검색>Rest API>책 검색하기 
# 헤더 : Authorization: KakaoAK ${REST_API_KEY}
#  ${REST_API_KEY} 의미는 '저 자리에 변수값을 치환하세요.' 의미라서 괄호와 달러키 다 포함하여 아래처럼 입력
# Authorization: KakaoAK 87fa270de879f3dd014b8779623884db 

session.headers.update({          
    "Authorization": "KakaoAK %s" % key
    })
```

```python

# 접속 정보 설정 # 오픈 api는 누구나 가져가라는 의미이므로 Referer가 필요하지 않다. 
# 검색>Daum검색>Rest API>책 검색하기 
# 헤더 : Authorization: KakaoAK ${REST_API_KEY}
#  ${REST_API_KEY} 의미는 '저 자리에 변수값을 치환하세요.' 의미라서 괄호와 달러키 다 포함하여 아래처럼 입력
# Authorization: KakaoAK 87fa270de879f3dd014b8779623884db 

session.headers.update({          
    "Authorization": "KakaoAK %s" % key
    })
```

```python
# 접속할 url
url = urlFmt.format(query = query, page = page, size = size)
url
```

```python
# 생성한 접속 객체를 활용하여 API에 접속
r = session.get(url)
```

```python
# 접속에 실패한 경우 
if r.status_code != 200 :
    msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)    # r.reason은 에러 메세지 원문임.
    raise Exception(msg)
```

```python
# 인코딩 형식 지정
r.encoding = "utf-8"
# 가져온 Json 문자열을 딕셔너리로 전환
data = json.loads(r.text) 
data
```

```python
# 필요한 부분만 추출하기 
# meta 는 응답관련 정보? : 현실세계를 대변하는 가상 공간이 메타 . 메타버스는 현실 세계에 영향을 준다. 그래서 아바타와는 개념이 다르다. 메타는 아래 documents를 설명하기 위한 부가정보다.documents : 응답결과  

# 결과 표 생성하기
df = DataFrame(data['documents'])
# 엑셀로 저장
df.to_excel('카카오 책 검색 결과%s.xlsx'%query)
df
```
