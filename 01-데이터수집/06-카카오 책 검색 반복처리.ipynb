```python
import requests
import json
from pandas import DataFrame # 딕셔너리를 포함한 리스트를 표 형태로 표기함
```

```python
# URL 기본 형식
# query 파라미터 : query는 변수 이름, 검색어 겟 방식이므로 ?query=검색어  그 외 더 찾고 싶은게 있다면 ?query=검색어&sort=latest?query=검색어&size=50 이런 방식으로 입력
urlFmt = "https://dapi.kakao.com/v3/search/book?query={query}&page={page}&size={size}"
```

```python
# 요청변수
query = "파이썬"
page = 1
size = 50
key = "87fa270de879f3dd014b8779623884db" # 카카오 개발자에서 발급받은 API 연동키
```

```python
# 접속 객체 생성
session = requests.Session()
session.headers.update({          
    "Authorization": "KakaoAK %s" % key
    })
```

```python

# 다음 페이지가 존재하는지 여부 -> 이 값이 False 인 동안은 반소한다
isEnd = False 
# 반복 

# 반복 수행 도중 추출되는 검색결과 리스트를 결합할 빈 리스트 준비
mylist= [] # 딕셔너리가 포함된 리스트 

# 1페이지 부터 시작
page = 1

while not isEnd :
    # -----------------------------------------------------------------
    # 이전 예제 코드 시작
    # ----------------------------------------------------------------
    # 접속할 url
    url = urlFmt.format(query = query, page = page, size = size)
    #url
    # 생성한 접속 객체를 활용하여 API에 접속
    r = session.get(url)

    # 접속에 실패한 경우 
    if r.status_code != 200 :
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)    # r.reason은 에러 메세지 원문임.
        raise Exception(msg)
    # 인코딩 형식 지정
    r.encoding = "utf-8"
    # 가져온 Json 문자열을 딕셔너리로 전환
    data = json.loads(r.text) 
    #data
    # -----------------------------------------------------------------
    # 이전 예제 코드 끝
    # ----------------------------------------------------------------
    # 필요한 부분만 추출하기 
    # meta 는 응답관련 정보? : 현실세계를 대변하는 가상 공간이 메타 . 메타버스는 현실 세계에 영향을 준다. 그래서 아바타와는 개념이 다르다. 메타는 아래 documents를 설명하기 위한 부가정보다.documents : 응답결과  

    # 검색 결과를 가족 있는 딕셔너리를 포함하는 리스트를 추출한다. 
    searchList = data['documents']

    # 추출한 리스트를 미리 준비한 빈 리스트에 추가
    mylist.extend(searchList)

    # 증감식에 해당하는 isEnd변수 추출
    isEnd = bool(data['meta']['is_end'])
    page += 1
    print("isEnd=", isEnd, ", 다음페이지=", page)

print("최종 검색 건수: %d" % len(mylist))
```

```python
# 검색 결과를 가족 있는 딕셔너리를 포함하는 리스트를 추출한다. 
searchList = data['documents']

# 추출한 리스트를 미리 준비한 빈 리스트에 추가
mylist.extend(searchList)

# 결과 표 생성하기
df = DataFrame(mylist)
# 엑셀로 저장
df.to_excel('카카오 책 검색 결과_최종%s.xlsx'%query)
df
```
