### 지도 객체에 마커 추가 
마커 = 지도에 위치 표시한거 
popup : 클릭시 나올 메세지 

주피터는 웹기반이라서, html로 마커를 수정한다. 

'lightgreen', 'darkgreen', 'darkblue', 'cadetblue', 'orange', 'lightred', 'darkred', 'green', 'blue', 'black', 'lightblue', 'white', 'lightgray', 'red', 'pink', 'beige', 'gray', 'purple', 'darkpurple'

```python
# 일반 마커

# 지도 객체 생성
map_osm1 = folium.Map(location=[37.4963462,126.9568865],zoom_start=15)
map_osm1 

# 마커 객체 생성
# 마커가 표시될 위치[위도,경도]
# 클릭 시 표시될 메세지, 아이콘 색상 
markar1 = folium.Marker([37.4963462,126.9568865],
                        popup='숭실대',
                        icon=folium.Icon(color='lightblue'))

markar1.add_to(map_osm1) # 마커 객체를 지도에 추가함
map_osm1
```

```python
# 사용자 지정 아이콘, HTML 팝업
# 지도 객체 생성
map_osm2 = folium.Map(location=[37.5004444, 127.0313861],zoom_start=15)

# HTML을 사용한 팝업 # No wrap = 엔터 치지마라 
popup_html = folium.Popup('<div style="white-space : nowrap"><h3>메가스터디 강남IT</h3>')

# 사용자 지정 아이콘 이미지 사용
# 온라인 상의 url 내 컴퓨터의 상의 상대/절대 경로 모두 사용 가능함. 
icon_imp = folium.features.CustomIcon('https://data.hossam.kr/D04/ho.png',icon_size=(50,50))

# 마커 객체 생성 
markar2 = folium.Marker([37.5004444, 127.0313861],
                        popup=popup_html,
                        icon=icon_imp)
markar2.add_to(map_osm2)

# 웹페이지 파일이 저장될 폴더 생성 
# not.path.exist 주어진 파일이 존재하면 넘어가고 없으면 만들어라 
if not path.exists('output') :
    mkdir('output')
map_osm2.save('output/map_osm2.html') #파일이 저장될 위치
map_osm2
```

```python
# 원형마커(범위지정)

# 지도 객체 생성
map_osm3 = folium.Map(location=[37.5004444, 127.0313861],zoom_start=15)

# 마커 객체 생성 
markar1 = folium.Marker([37.5004444, 127.0313861],
                        icon=folium.Icon(color='orange'))
markar1 = folium.CircleMarker([37.5004444, 127.0313861],
                              radius=100, # 원크기
                              color='#3186cc',  # 선 색상
                              fill_color='#3186cc') # 면 색상 

markar2.add_to(map_osm3)
markar1.add_to(map_osm3)

# 마커 객체 생성 
markar3 = folium.Marker([37.5004444, 127.0313861],
                        icon=folium.Icon(color='orange'))
map_osm3
```

예제) 서울의 고등학교 분포 확인하기 (공공데이터 포털)

```python
xlsx = read_excel("https://data.hossam.kr/D04/school.xlsx")
xlsx
```

```python
# 데이터 전처리 (사용할 필드만 추출)
df = xlsx.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
df
```

```python
# '서울'시 '고등학교'만 추출 
# # LIKE 연산 -> 컬럼이름.str.contains('검색어')
df2 = df.query("학교급구분 == '고등학교' and 소재지도로명주소.str.contains('서울')")
df2
```

```python
# zoom_start: 배율 1~22 (여기서는 출력 안함)
map_osm = folium.Map(location=[37.566651, 126.978428], zoom_start=12)

html = "<font color='green' style='white-space: nowrap'><b>%s</b></font>"

# 데이터프레임의 행 수만큼 반복하면서 마커 생성
for i in df2.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = df2.loc[i, '학교명']
    lat = df2.loc[i, '위도']
    lng = df2.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
    popup_html = folium.Popup(html % name, parse_html=False) # False로 줘야 name이 출력된다. 아니면 html이 모두 출력된다. 
    marker = folium.Marker([lat,lng], popup=popup_html)
    marker.add_to(map_osm)

map_osm
```
