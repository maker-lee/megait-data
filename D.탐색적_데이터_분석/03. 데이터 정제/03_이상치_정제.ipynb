### 찾아낸 이상치

1. 상자그림을 통해 검출된 국어, 수학, 과학 점수 (동그라미)
2. 분석가의 주관에 따른 영어 점수 (`> 100`)


##### 사분위수 
- 데이터 표본을 4개의 동일한 부분으로 나눈 값
- 사분위수를 사용하여 데이터 집합의 범위와 중심 위치를 신속하게 평가할 수 있다. 
- | 1사분위 수(Q1)     | 데이터의 하위 25%가 이 값보다 작거나 같음.             |
- | 2사분위 수(Q2)     | 중위수 데이터의 50%가 이 값보다 작거나 같음. (=중앙값) |
- | 3사분위 수(Q3)     | 데이터의 하위 75%가 이 값보다 작거나 같음.(=상위 25%)  |
- 1사분위 수 $ Q1 = ((총도수 + 1)*0.25)+1 $
- 3사분위 수 $ Q3 = ((총도수 + 1)*0.75)+1 $
- $ IQR = Q3 - Q1 $사분위의 정상 범위인 Q1과 Q3 사이를 의미하며 사분범위라고 한다. 



### 이상치를 결측치로 변경 (구현한 함수 호출)

```python
re_df = replaceOutlier(re_df, ["국어", "수학", "과학"])
re_df
```

### 영어 점수 처리

분석가에 주관에 따른 이상치 선정 결과

```python
re_df.loc[re_df["영어"] > 100, "영어"] = np.nan # 영어에서 100점 이상은 이상치이므로 np.nan으로 처리한다. 
re_df
```

### 변경된 결측치에 대한 처리

```python
final_df = replaceMissingValue(re_df) # MissingValue(결측치) 값을 모두 mean 값으로 대체해라
final_df
```
