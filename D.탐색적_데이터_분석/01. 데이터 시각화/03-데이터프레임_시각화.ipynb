# 데이터프레임 시각화

데이터프레임이 내부적으로 pyplot 객체를 초기화 하기 때문에 plt.figure()를 호출할 필요가 없다. 
데이터프레임 그대로를 (분해하지 않고) 그래프로 옮기는 방법 . 객체.plot()하면 가능함.

```python
from pandas import read_excel
from matplotlib import pyplot as plt
```

기본 세팅

```python
plt.rcParams['font.family'] = "Malgun Gothic"
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (20,8)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx", index_col='년도' )
df
```

## Line Plot 

1. 특정 컬럼에 대한 시각화
- 기본적으로 dataframe의 컬럼 이름이 label이 되지만, plot() 메서드의 label 파라미터를 설정하여 다른 이름으로 변경할 수 있다.

```python
df['발생건수'].plot(label='교통사고')
plt.grid()
plt.legend()
plt.title('연도별 교통사고 발생건수')
plt.show()
plt.close()
```

2. 2개 이상 컬럼에 대한 시각화

```python
df.filter(['사망자수','부상자수']).plot()
plt.grid()
plt.legend()
plt.title('교통사고 사망자수, 부상자수')
plt.show()
plt.close()
```

3. 전체컬럼에 대한 시각화

```python

df.plot()
plt.grid()
plt.legend()
plt.title('년도별 교통사고')
plt.ylabel('교통사고 수') # 데이터 범주 이름은 column이름이다. 바꾸고 싶으면 label을 따로 줘야 한다.
plt.show()
plt.close()
```

## Bar Plot
1 특정 컬럼에 대한 시각화

```python
df['발생건수'].plot.bar(rot=0)
plt.grid()
plt.legend()
plt.title('년도별 교통사고 발생건수')
plt.ylim(200000,240000)
plt.show()
plt.close()
```

2개 이상 컬럼에 대한 시각화

```python
df.filter(['발생건수','부상자수']).plot.bar(rot=45)# x축 이름이 세로로 서서 각도를 rot = 0으로 줘야 한다. 
plt.grid()
plt.legend()
plt.title('년도별 교통사고 발생건수')
plt.ylim(150000, 370000)
plt.show()
plt.close()
```

3. 전체 컬럼에 대한 시각화

```python
df.plot.bar(rot=0)
plt.grid()
plt.legend()
plt.title('년도별 교통사고 발생건수')
plt.show()
plt.close()
```

## 산점도 그래프

```python
df.plot.scatter(x='발생건수',y='부상자수',color='#a3506d')
plt.grid()
plt.show()
plt.close()
```

## Pie그래프
- dataframe의 index가 파이그래프의 한 조각이 된다.
- 파이 그래프는 컬럼 단위로 해야 한다. 

```python
df['사망자수'].plot.pie(autopct='%0.1f%%')
plt.show()
plt.close()
```
