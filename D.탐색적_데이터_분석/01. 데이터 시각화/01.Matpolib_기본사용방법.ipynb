# Matplotlib 기본 사용 방법
## 01. Matplotlib 개요
- Matplotlib는 파이썬에서 자료를 그래프로 시각화하는 패키지이다.
- Matplitlib에 포함된 pyplot이라는 객체를 통해 다양한 시각화 기능을 제공함.

1. 표현 가능한 그래프 종류
- 선 : 라인 플롯(line plot)
- 막대 : 바 차트(bar chart)
- 상자그림 : 박스 플롯(box plot)
- 원 : 파이 차트(pie chart)
- 히스토그램(histogram)
- 산점도 : 스캐터 플롯(scatter plot)


```python
from matplotlib import pyplot as plt
```

2. 라인 플롯
- 가장 간단한 플롯은 선을 그리는 라인 플롯이다
- 라인 플롯은 데이터가 시간, 순서 등에 따라 어떻게 변화하는지 보여주기 우해 사용한다.
- 명령은 pyplot 객체의 plot() 메소드를 사용한다.

1. 기본 그리기
- plot() 메소드에 전달되는 리스트나 numpy 배열의 인덱스가 x축이 되고 값이 y축이 된다

```python
plt.figure() # 그래프 시작 선언
plt.title('MY Plot') # 제목
plt.plot([1,3,5,7,9]) # 그래프 그리기 (x축은 인덱스)
plt.xlabel('x') # x축 제목
plt.ylabel('y') # y축 제목
plt.show() # 그래프 화면 표시
plt.close() # 그래프 종료
```

3. 한글 사용ㅇ하기
- 한글을 사용하려면 pyplot객체에 한글 폰트를 설정해야 한다.
- 한글 인식을 못함...............바보임.............. 
- 윈도우는 TTF로 폰트 다운로드 -> 그런데 폰트이름이 실제 설치된 이름이 아니다 
- Malgun Gothic 맑은고딕 = 윈도우 전용
- AppleGothic 애플고딕 = Mac 전용 

1. 컴퓨터에 설치되어 있는 글꼴 다 확인해보기

```python
import os
import sys
from matplotlib import font_manager
```

```python
# 시스템 글꼴 목록을 리스트로 가져오기
font_list = font_manager.findSystemFonts()
# 이름순으로 정렬함
font_list.sort()

# 각 폰트 파일의 경로에 대한 반복
for font_path in font_list :
    # 폰트 파일의 경로를 사용하여 폰트 속성 가져오기
    fp = font_manager.FontProperties(fname=font_path)
    #print(fp)

    try :
        font_name = fp.get_name()
        print(font_name)
    except :
        continue
```

2. 글꼴을 적용하기 위한 그래프 전역 설정
- 이 설정은 같은 내용이 재설정되기 전까지 모든 그래프에 공통 적용된다

```python
# 폰트 설정
plt.rcParams['font.family'] = "Malgun Gothic"
# 글자 크기
plt.rcParams['font.size'] = 12
# 그래프 크기 설정(inch) inch 의 감이 없으니까 걍 inch [10,5]로 잡고 늘렸다가 줄였다가 하세욤
plt.rcParams['figure.figsize'] = (10,5)
# 유니코드에서 음수 부호 설정 (안하면 D로 표기됨)
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 한글 테스트
plt.figure()
plt.title('한글이 될까용')
plt.xlabel('x축')
plt.ylabel('y축')
plt.plot([-20,-10,0,10,20,30,40])
plt.show()
plt.close()
```

4. 각 영역별로 개별 폰트 지정

```python
# 컬러는 구글에서 colot picker 검색 HEX 사용
# windows font
font1 = {'family': 'Batang', 'size': 24, 'color': '#2aa2c9', 'weight': 'bold'}
font2 = {'family': 'Gulim', 'size': 12, 'color': '#0066ff'}
font3 = {'family': 'Gulim', 'size': 12, 'color':  '#ff6600', 'weight': 'light'}

plt.figure()
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
plt.close()
```

4. 스타일 지정
플롯 명령어는 보는 사람이 그림을 더 알아보기 쉽게 하기 위해 다양한 스타일을 지원한다
plot() 명령어에서는 다음과 같이 추가 문자열 인수를 사용하여 스타일을 지원한다.

```python
'''
1) 선스타일 linestyle
- : 실선
-- : 대시선
: : 점선
-. : 대시-점선
'''
```

```python
plt.figure()
plt.title('스타일 지정')
plt.plot([10,20,30,40,50],[1,4,9,16,25],linestyle=":")
plt.grid() # 그리드(격자) 표시
plt.show()
plt.close()
```

2. marker 마커
- 데이터 위치를 나타내는 기호를 마커라고 한다. 마커의 종류는 아래와 같다.

마커 문자열	의미
.	point marker
,	pixel marker
o	circle marker
v	triangle_down marker
^	triangle_up marker
<	triangle_left marker
>	triangle_right marker
1	tri_down marker
2	tri_up marker
3	tri_left marker
4	tri_right marker
s	square marker
p	pentagon marker
*	star marker
h	hexagon1 marker
H	hexagon2 marker
+	plus marker
x	x marker
D	diamond marker
d	thin_diamond marker

```python
plt.figure()
plt.title('마커 연습')
plt.plot([10,20,30,40,50],[1,4,9,16,25],linestyle="--",marker="^")
plt.show()
plt.close()
```

3. 기타 스타일

스타일 문자열	약자	의미

linestyle	ls	선 스타일
linewidth	lw	선 굵기
marker		마커 종류
markersize	ms	마커 크기

```python
plt.figure()
plt.title('스타일 지정')
plt.plot([10,20,30,40,50],[1,4,9,16,25],linestyle="--",linewidth=3,marker="^",markersize='20',markerfacecolor='#fffff0',markeredgecolor='#000000',markeredgewidth=2)
plt.show()
plt.close()
```

5. 축설정
1. 범위지정
x축의 범위(xlim)와 y축의 범위(ylim)을 설정

```python
plt.figure()
plt.title('축 설정')
plt.plot([10,20,30,40,50],[1,4,9,16,25])
plt.xlim(0,60) # 범위
plt.ylim(-10,30)
plt.show()
plt.close()
```

2. 각 축의 표시 내용 설정

```python
y_valuse = list(range(70,101,5))
y_ticks = []
for i in y_valuse :
    y_ticks.append('%d점' % i)
y_ticks
```

```python
plt.figure()
plt.title('학년별 평균 점수 변화')
plt.plot([1,2,3,4],[98,58,76,88])
# x축에 표시될 내용 설정 -> 좌표값 리스트와 표시 내용 리스트
plt.xticks([1,2,3,4],['1학년','2학년','3학년','4학년'])
plt.yticks(y_valuse,y_ticks)

plt.gray()
plt.show()
plt.close()
```

6 겹쳐 그리기
pyplot은 figure() 메소드를 호출한 이후부터 show(), close() 메서드를 호출하기 전까지의 코드들을 겹쳐 표현할 수 있다. 

```python
plt.figure()
plt.title('여러 그래프 한 번에 그리기')
plt.plot([1,2,3,4,5,6,7])
plt.plot([1,2,5,6,8,13,14])
plt.show()
plt.close()
```

7. 범례 (legend)
여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지 보여주기 위해서 legend 명령으로 범례를 추가한다. 범례의 위치는 자동이지만 수동으로 하기 위해서는 loc를 사용한다. 인수에는 문자열 혹은 숫자가 들어간다.   lable=범주 는 legend(loc=위치)를 호출해야지 표기가 가능하다. 

loc 문자열	숫자
best	0
upper right	1
upper left	2
lower left	3
lower right	4
right	5
center left	6
center right	7
lower center	8
upper center	9
center	10

```python
plt.figure()
plt.title('범례 사용')
plt.plot([1,2,3,4,5,6,7],label='1차 방정식')
plt.plot([1,2,5,6,8,13,14],label='2차 방정식')
plt.legend(loc='lower right')
plt.show()
plt.close()
```
