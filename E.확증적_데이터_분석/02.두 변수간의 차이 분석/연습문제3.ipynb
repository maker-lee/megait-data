## 문제 1

유아들을 대상으로 세 가지 읽는 방법을 비교,실험하여 다음과 같은 독서평가 점수자료를 얻었다.

세 방법에 통계적으로 유의한 차이가 있는지 검정하라.

![res/q31.png](res/q31.png)

> https://data.hossam.kr/E02/reading.xlsx

```python
df = read_excel("https://data.hossam.kr/E02/reading.xlsx")
df
```

```python
독서방법_종류 = list(set(df['독서방법'].values))
독서방법_종류
```

```python
# 데이터 확인
독서방법_종류 = list(set(list(df['독서방법'])))

# 데이터 전처리
독서평가_점수 = df.copy()
독서평가_점수['독서방법'] = 독서평가_점수['독서방법'].map({독서방법_종류[0]:0,독서방법_종류[1]:1,독서방법_종류[2]:2})
독서평가_점수
```

```python
# 정규성, 등분산성, 독립성 확인
all_test(독서평가_점수['독서방법'],독서평가_점수['점수'])
```

데이터가 적어서 굳이 all_test할 필요는 없다. 

```python
# 일원분산분석 수행
lm = ols('점수 ~ C(독서방법)', data=독서평가_점수).fit()
anova_lm(lm)
```

p-value가 0.05보다 크므로 귀무가설을 기각할 수 없다. 즉, 통계적으로 유의미한 차이는 발견하지 못했다.

분산분석을 하여 효소 함량에 따라 토마토 모종 성장에 차이가 나는지 검정하라.

```python
df1 = read_excel("https://data.hossam.kr/E02/tomato.xlsx")
df1
```

숫자로 하면 큰 차이가 나서 반드시 라벨링 해줘야한다. 그리고 라벨링은 0으로 하면 에러나니까 1부터해라

```python
df1.info()
```

```python
df1['효소함량']=df1['효소함량'].map({0:1,1000:1,독서방법_종류[2]:2}) # 일일히 치기 시른뎅 
```

```python
all_test(df1['효소함량'],df1['모종성장'])
```

이 자료를 이용하여 두 가지 교육방법에 대한 국어와 영어의 효과가 서로 다르다고 할 수 있는지 검정하라. (분석 목적에 맞는 데이터 재배치 과정이 필요합니다.)

### 정답풀이

교육방법에 따라 국어와 영어의 상관을 알아내는게 문제

```python
교육_국어 = 교육.filter(['교육방법','국어'])
교육_국어['교과목'] = '국어'
교육_국어.columns=['교육방법','점수','교과목']
# 교육_국어.rename(columns={'국어':'점수},inplace=True)
교육_국어
```

```python
교육_영어 = 교육.filter(['교육방법','영어'])
교육_영어['교과목'] = '영어'
교육_영어.columns=['교육방법','점수','교과목']
교육_영어
```

```python
교육_가공=concat([교육_국어,교육_영어])
교육_가공
```

```python
교육_가공['교과목'] = 교육_가공['교과목'].map({'국어': 1, '영어': 2})
df3 = 교육_가공.copy()
df3
```

```python
formula = '점수 ~ C(교육방법) * C(교과목)'
lm = ols(formula, 교육_가공).fit()
anova = anova_lm(lm)
anova
```

교육방법에 따라 유의미한 차이는 없다.
