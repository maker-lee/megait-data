## 문제 1

다음 자료는 야구팀의 홈런수와 타율에 관한 자료이다. 홈런수와 평균타율이 상관관계가 있는지 그 관련성을 분석하라.

!()[res/q01.png]

> https://data.hossam.kr/E03/baseball.xlsx

```python
df = read_excel("https://data.hossam.kr/E03/baseball.xlsx")
df
```

```python
sb.regplot(data=df,x='홈런수',y='평균타율')
plt.grid()
plt.show()
plt.close()
```

```python
corr = df.corr(method='pearson')
corr
```

```python
plt.rcParams['figure.figsize'] = (6,4)
sb.heatmap(corr,
           annot = True, # 실제 값 화면에 나타내기
           cmap = "Blues", # 색상
           vmin= -1, vmax= 1, # 색상 차트 영역
           linewidths=0.5) # 구분선 

plt.show()
plt.close()
```

```python
stats.pearsonr(df['홈런수'],df['평균타율'])
```

p-value가 유의수준 0.05보다 크므로 귀무가설을 기각할 수 없다.

## 문제 2

어떤 원료의 품질특성($x$)과 이 원료를 사용하여 만든 제품의 품질특성($y$)에 관한 조사를
한 결과 다음 데이터를 얻었다.

이들 간 상관관계를 구하고 상관 유무 검정을 유의수준 `5%`에서 실시하라.

!()[res/q02.png]

> https://data.hossam.kr/E03/product.xlsx

```python
df1 = read_excel("https://data.hossam.kr/E03/product.xlsx")
df1
```

```python
sb.regplot(data=df1,x='x',y='y')
plt.grid()
plt.show()
plt.close()
```

```python
corr = df1.corr(method='pearson')
corr

plt.rcParams['figure.figsize'] = (6,4)
sb.heatmap(corr,
           annot = True, # 실제 값 화면에 나타내기
           cmap = "Blues", # 색상
           vmin= -1, vmax= 1, # 색상 차트 영역
           linewidths=0.5) # 구분선 
plt.show()
plt.close()
```

```python
r, p = stats.pearsonr(df1['x'],df1['y'])
"상관계수: %f, p-value: %f" % (r, p)
```

p-value가 유의수준 0.05보다 작으므로 귀무가설을 기각하고 대립가설을 채택한다. 그러므로 품질특성($x$)과 이 원료를 사용하여 만든 제품의 품질특성($y$)는 상관이 있다. 

## 문제 3

다음은 10명의 학생들에 대한 공부 시간과 성적 데이터이다. 분석하라. 네?

!()[res/q03.png]

> https://data.hossam.kr/E03/study_time.xlsx

```python
df2 = read_excel("https://data.hossam.kr/E03/study_time.xlsx")
df2
```

```python
sb.regplot(data=df2,x='공부시간',y='성적')
plt.grid()
plt.show()
plt.close()
```

```python
corr = df2.corr(method='pearson')
corr

plt.rcParams['figure.figsize'] = (6,4)
sb.heatmap(corr,
           annot = True, # 실제 값 화면에 나타내기
           cmap = "Blues", # 색상
           vmin= -1, vmax= 1, # 색상 차트 영역
           linewidths=0.5) # 구분선 
plt.show()
plt.close()
```

```python
r, p = stats.pearsonr(df2['공부시간'],df2['성적'])
"상관계수: %f, p-value: %f" % (r, p)
```

p-value가 유의수준 0.05보다 작으므로 귀무가설을 기각하고 대립가설을 채택한다. 그러므로 공부시간과 성적에는 약한 양의 상관관계가 있다.
